### Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
from datetime import datetime
```

### Find and load lux data

```python
dfs = []
lux_files = []
measurement_files = []
for root, dirs, files in os.walk("../data/lux", topdown=False):
    for name in files:
        if '.csv' in name:
            lux_files.append(os.path.join(root, name))
lux_files = np.sort(lux_files)
for lux_file in lux_files:
    df = pd.read_csv(lux_file,header=None)
    df.columns = ['lux','year','month','day','hour','minute','second']
    dfs.append(df)
lux_df = pd.concat(dfs)
```

### Find and load measurement data

```python
dfs = []
measurement_files = []
for root, dirs, files in os.walk("../data/measurements", topdown=False):
    for name in files:
        if '.csv' in name:
            measurement_files.append(os.path.join(root, name))
measurement_files = np.sort(measurement_files)
for measurement_file in measurement_files:
    df = pd.read_csv(measurement_file,header=None)
    df.columns = ['pressure','height','pressureTemperature','humidity','humidityTemperature',
                  'year','month','day','hour','minute','second']
    dfs.append(df)
measure_df = pd.concat(dfs)
measure_df['month'] = measure_df['month'].astype('str')
measure_df['day'] = measure_df['day'].astype('str')
measure_df['month'] = measure_df['month'].str.zfill(2)
measure_df['day'] = measure_df['day'].str.zfill(2)
```

### Convert to datetime object

```python
measure_df['datetime_str'] = measure_df['month'].astype(str) + '/' + \
                             measure_df['day'].astype('str') + '/' + \
                             measure_df['year'].astype('str') + ' ' + \
                             measure_df['hour'].astype('str') + ':' + \
                             measure_df['minute'].astype('str') + ':' + \
                             measure_df['second'].astype('str')
measure_df['datetime_str'].values
datetime_objects = []
for datetime_str in measure_df['datetime_str'].values:
    datetime_object = datetime.strptime(datetime_str, '%m/%d/%Y %H:%M:%S')
    datetime_objects.append(datetime_object)
measure_df['datetime'] = datetime_objects

lux_df['datetime_str'] = lux_df['month'].astype(str) + '/' + \
                             lux_df['day'].astype('str') + '/' + \
                             lux_df['year'].astype('str') + ' ' + \
                             lux_df['hour'].astype('str') + ':' + \
                             lux_df['minute'].astype('str') + ':' + \
                             lux_df['second'].astype('str')
lux_df['datetime_str'].values
datetime_objects = []
for datetime_str in lux_df['datetime_str'].values:
    datetime_object = datetime.strptime(datetime_str, '%m/%d/%Y %H:%M:%S')
    datetime_objects.append(datetime_object)
lux_df['datetime'] = datetime_objects
```

### Plot Lux

```python
plt.plot(lux_df['datetime'],lux_df['lux'],'r')
plt.xticks(lux_df['datetime'][::16000],rotation=90);
plt.grid();
```

### Plot Barometric Pressure

```python
measure_df.columns
plt.plot(measure_df['datetime'],measure_df['pressure'],'r')
plt.xticks(measure_df['datetime'][::16000],rotation=90);
plt.grid();
```

### Plot Temperature

```python
plt.plot(measure_df['datetime'],measure_df['humidityTemperature'],'r')
plt.xticks(measure_df['datetime'][::16000],rotation=90);
plt.grid();
```

### Plot Humidity

```python
plt.plot(measure_df['datetime'],measure_df['humidity'],'r')
plt.xticks(measure_df['datetime'][::16000],rotation=90);
plt.grid();
```
