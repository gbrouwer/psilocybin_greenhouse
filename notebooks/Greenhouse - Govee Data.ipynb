### Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
from datetime import datetime
```

### Load Data

```python
df = pd.read_csv('../data/govee/humidity_export_202302111158.csv')
df['datetime'] = pd.to_datetime(df['Timestamp'].values)
df.columns
```

### Plot Temperature

```python
plt.plot(df['datetime'],df['Temperature_Fahrenheit'],'r')
plt.xticks(df['datetime'][::40],rotation=90);
plt.grid();
```

### Plot Humidity

```python
plt.plot(df['datetime'],df['Relative_Humidity'],'r')
plt.xticks(df['datetime'][::40],rotation=90);
plt.grid();
```
